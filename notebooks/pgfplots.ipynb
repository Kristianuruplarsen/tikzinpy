```python
import tikzinpy as tip
from tikzinpy import pgfplots
from tikzinpy import bshow #build and show

import seaborn as sns 
import numpy as np 
import matplotlib.pyplot as plt 

# This is just some data to test with
real_data = sns.load_dataset('iris')

data_x = np.linspace(-2, 6, 100)
data_e = np.random.uniform(0,1, size = 100)
data_y = 0.5*data_x + data_e
```

```python
base = tip.tikzBase()
ax = pgfplots.pgfAxis()

ax += pgfplots.pgfscatterplot(data_x, data_y)

base += ax
```

```python
bshow(base, 'test/', 'pgfscatter')
```
